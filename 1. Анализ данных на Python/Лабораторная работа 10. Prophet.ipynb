```python
from prophet import Prophet
```

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
df_1 = pd.read_csv('Lect_23/benzak_01.csv', sep=';', )
df_1.head()
```

```python
plt.figure(figsize=(18, 8))
plt.plot(df_1['y'])
```

```python
horizon_ = 60
```

```python
m_1 = Prophet()
m_1.fit(df_1)
```

```python
forecast_1 = m_1.make_future_dataframe(periods=horizon_)
forecast_1.tail()
```

```python
forecast_1 = m_1.predict(forecast_1)
forecast_1[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m_1.plot(forecast_1)
```
